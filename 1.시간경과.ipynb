[1] : 시간의 흐름에 따라 매출의 추이는 어떻게 달라지는가

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('data/ecommerce_data_final.csv')
```

매출 변화

```python
# 일자별 총 매출
df.groupby('date_ymd')[['amount']].sum().reset_index()
```

```python
amount_by_date = df.groupby('date_ymd')[['amount']].sum().reset_index()

fig = px.line(data_frame = amount_by_date, x = 'date_ymd', y = 'amount')
fig.show()
```

주문고객수 변화

```python
# 일자별 총 매출
customer_count_by_date = df.groupby('date_ymd')[['CustomerID']].nunique().reset_index()
customer_count_by_date.rename(columns={'CustomerID' : 'customer_count'}, inplace=True)

customer_count_by_date
```

```python
fig = px.line(data_frame = customer_count_by_date, x = 'date_ymd', y = 'customer_count')
fig.show()
```

주문단가 = 총매출 / 주문건수

```python
amount_by_date.head()
```

```python
invoice_count_by_date = df.groupby('date_ymd')[['InvoiceNo']].nunique().reset_index()
invoice_count_by_date.rename(columns = {'InvoiceNo' : 'invoice_count'}, inplace = True)

invoice_count_by_date
```

```python
fig = px.line(data_frame = invoice_count_by_date, x = 'date_ymd', y = 'invoice_count')
fig.show()
```

```python
invoice_amount = pd.merge(amount_by_date, invoice_count_by_date, on = 'date_ymd')

# 주문단가 컬럼 생성
invoice_amount['amount_per_invoice'] = invoice_amount['amount'] / invoice_amount['invoice_count']
invoice_amount.head()
```

```python
fig = px.line(data_frame = invoice_amount, x = 'date_ymd', y = 'amount_per_invoice')
fig.show()
```

# 전처리
1. 결측치 처리
2. 데이터 타입 변경
3. 새로운 컬럼 추가
4. 불필요한 데이터 삭제

---
### 1. 결측치 처리
CustomerID가 없는 데이터의 경우 분석에 의미가 없으므로 결측치 데이터 삭제

```python
df.dropna(subset = ['CustomerID'], inplace = True)

df.info()
```

---
### 2. 데이터 타입 변경
1) InvoiceDate 날짜 타입으로 변경
2) CustomerID 문자형으로 변경

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format = '%m/%d/%Y %H:%M')
df['CustomerID'] = df['CustomerID'].astype(int).astype(str)

df.info()
```

```python
df.head()
```

---
### 3. 새로운 컬럼 추가
1. InvoiceDate에서 연도와 날짜 분리하여 새로운 컬럼으로 추가
2. Quantity * UnitPrice	의 amount(매출) 컬럼 추가

```python
import datetime as dt

df['date_ymd'] = df['InvoiceDate'].dt.date.astype('datetime64[ns]')
df['year'] = df['InvoiceDate'].dt.year

df.head()
```

```python
df['amount'] = df['Quantity'] * df['UnitPrice']

df.head()
```

---
### 4. 불필요한 데이터 삭제
성공적인 주문의 경우만 분석에 활용하기 위해 취소 주문 데이터 삭제

```python
df[df['Quantity'] < 0]
```

```python
df = df[~df['Quantity'] < 0].reset_index(drop = True)

df.head()
```
